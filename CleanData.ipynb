```python
file = "Resources/streaming_movies.csv"

streaming = pd.read_csv(file)

streaming.head()
```

```python
streaming_clean = streaming.dropna()

print(len(streaming_clean))

streaming_clean = streaming_clean.loc[streaming_clean["Country"] == "United States"]

print(len(streaming_clean))

streaming_clean.head()
```

```python
merged_df = pd.merge(movies_clean, streaming_clean, on="Title", how="outer")

merged_df = merged_df.dropna()

merged_df.head()
```

```python
merged_df = merged_df.rename(columns= {"type": "TV Show/Movie", "director": "Director", "cast": "Cast", "date_added": "Date Added to Netflix", "release_year": "Year Released", "rating": "Rating", "duration": "Duration"})

merged_df.head()
```

```python
merged_df = merged_df.drop(columns =["show_id", "Type", "Runtime", "Country", "Age", "Year", "ID", "Unnamed: 0", "Directors", "TV Show/Movie"])

merged_df = merged_df.rename(columns={"listed_in": "Netflix Categories", "country": "Country", "description": "Description", "IMDb": "IMDb Rating", "Rotten Tomatoes": "Rotten Tomatoes Rating"})

merged_df.head()
```

```python
merged_df['Year Released'] = merged_df['Year Released'].astype(int)
merged_df["Rotten Tomatoes Rating"] = merged_df["Rotten Tomatoes Rating"].str.rstrip("%")
merged_df['Rotten Tomatoes Rating'] = merged_df['Rotten Tomatoes Rating'].astype(float)
merged_df.dtypes
```

```python
merged_df["IMDb/Rotten Tomatoes Combined Rating"] = ((merged_df["IMDb Rating"] / 10 * 100 ) + (merged_df["Rotten Tomatoes Rating"])) / 2
merged_df.head()
```

```python
merged_df["Rotten Tomatoes Rating"] = merged_df["Rotten Tomatoes Rating"].astype(str) + '%'
merged_df["IMDb/Rotten Tomatoes Combined Rating"] = merged_df["IMDb/Rotten Tomatoes Combined Rating"].astype(str) + '%'
merged_df.head()
```

```python
merged_df = merged_df[["Title", "Director", "Cast", "Country", "Date Added to Netflix", "Year Released", "Rating", "Duration", "Netflix Categories", "Description", "IMDb Rating", "Rotten Tomatoes Rating", "IMDb/Rotten Tomatoes Combined Rating", "Netflix", "Hulu", "Prime Video", "Disney+", "Genres", "Language"]]
merged_df.head()
```
